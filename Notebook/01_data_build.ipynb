## Build Data

```python
def merge_csv_files(
    folder_path: str,
    output_file: str,
    datetime_col: str,
    sort_by_datetime: bool = True
):

    all_files = glob.glob(os.path.join(folder_path, "*.csv"))
    print(f"Found {len(all_files)} files. Starting merge...")

    if not all_files:
        raise FileNotFoundError(f"No CSV files found in {folder_path}")

    df_list = []

    for filename in all_files:
        df = pd.read_csv(filename, index_col=None, header=0)
        df_list.append(df)

    full_df = pd.concat(df_list, axis=0, ignore_index=True)

    if sort_by_datetime and datetime_col in full_df.columns:
        full_df[datetime_col] = pd.to_datetime(full_df[datetime_col], errors="coerce")
        full_df = full_df.sort_values(by=datetime_col)

    full_df.to_csv(output_file, index=False)

    print(f"Success! Combined file saved as '{output_file}'")
    print(f"Total rows: {len(full_df)}")

    return full_df.head()
```

```python
merge_csv_files(
    folder_path="../aqData",
    output_file="aq_data_full.csv",
    datetime_col="datetimeUtc",
)
```

```python
merge_csv_files(
    folder_path="../weatherData",
    output_file="weather_data_full.csv",
    datetime_col="datetime",
)
```
