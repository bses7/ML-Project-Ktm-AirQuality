```python
def missing_statistics(df):    
    statitics = pd.DataFrame(df.isnull().sum()).reset_index()
    statitics.columns=['COLUMN NAME',"MISSING VALUES"]
    statitics['TOTAL ROWS'] = df.shape[0]
    statitics['% MISSING'] = round((statitics['MISSING VALUES']/statitics['TOTAL ROWS'])*100,2)
    return statitics[["COLUMN NAME","% MISSING"]]
```

## What to do

DROP SEVERRISK

- Cannot reliably guess (impute) 90% of a column. "Severe Risk" is usually a calculation for thunderstorms/tornadoes. In Kathmandu, this isn't the primary driver for PM2.5. Keeping it would just add "noise" in the model.

VISIBILITY TIME SERIES APPROACH

- Visibility doesn't jump from 0km to 15km in one second. It changes gradually as fog or smog lifts. Linear interpolation looks at the value before and after the gap and draws a straight line between them.

FFILL WINDGUST

- It’s just one hour. We can safely assume the wind gust was similar to the hour before it.


```python
cols_to_drop = ['severerisk']
weather_clean = weather_clean.drop(columns=[c for c in cols_to_drop if c in weather_clean.columns])
```

## Feature Engineering


```python
def add_temporal_features(df):
    df['hour'] = df['timestamp'].dt.hour
    df['month'] = df['timestamp'].dt.month
    df['day_of_week'] = df['timestamp'].dt.dayofweek
    
    # 2. Season Mapping (The Kathmandu Cycle)
    def get_season(month):
        if month in [12, 1, 2]: return 'Winter'        # Cold, Inversion Layer
        elif month in [3, 4, 5]: return 'Pre-Monsoon'  # Dusty, Windy
        elif month in [6, 7, 8, 9]: return 'Monsoon'   # Clean, Rainy
        else: return 'Post-Monsoon'                    # Festivals, Harvest
    
    df['season'] = df['month'].apply(get_season)
    
    # 3. Brick Kiln Operation (Jan - May)
    df['brick_kiln_active'] = df['month'].apply(lambda x: 1 if 1 <= x <= 5 else 0)
    
    # 4. Cyclical Encoding for Month & Hour
    # This tells the model that Hour 23 is next to Hour 0, and Month 12 is next to Month 1
    df['month_sin'] = np.sin(2 * np.pi * df['month']/12)
    df['month_cos'] = np.cos(2 * np.pi * df['month']/12)
    df['hour_sin'] = np.sin(2 * np.pi * df['hour']/24)
    df['hour_cos'] = np.cos(2 * np.pi * df['hour']/24)
    
    # 5. Traffic/Rush Hour Logic (Kathmandu specific)
    df['is_rush_hour'] = df['hour'].apply(lambda x: 1 if (8 <= x <= 11) or (16 <= x <= 20) else 0)
    
    # 6. Nepal Weekend (Saturday is Day 5)
    df['is_weekend'] = df['day_of_week'].apply(lambda x: 1 if x == 5 else 0)
    
    return df
```

```python
def add_lag_features(df):
    # --- PM2.5 Momentum ---
    df['pm25_lag_1h'] = df['pm25'].shift(1)
    df['pm25_lag_2h'] = df['pm25'].shift(2)
    df['pm25_rolling_6h'] = df['pm25'].rolling(window=6).mean()
    df['pm25_rolling_24h'] = df['pm25'].rolling(window=24).mean()
    
    # --- Weather Interaction Lags ---
    # Rain 3 hours ago (Washing effect)
    df['precip_lag_3h'] = df['precip'].shift(3)
    
    # Visibility 3 hours ago (Clarity trend)
    df['visibility_lag_3h'] = df['visibility'].shift(3)
    
    # Temperature Change (Inversion trigger)
    # Difference between current temp and 6 hours ago
    df['temp_diff_6h'] = df['temp'] - df['temp'].shift(6)

    return df
```

### The "Strategic Fix" for the Model

When we move to training, we will use three specific techniques to make sure the model doesn't ignore that 7%:

Class Weighting: We tell the model: "Every time you miss a Hazardous hour, it's 10 times worse than missing a Safe hour." (In code: class_weight='balanced').

Precision-Recall Focus: We will ignore "Accuracy" and focus only on Recall (the percentage of hazardous days we successfully caught).

Threshold Tuning: Instead of the model needing a 50% probability to warn of a hazard, we might set it to 20%. It's better to be safe than sorry!
